```python
# lambda arguments: expression

l_add = lambda x, y: x + y
print(l_add(2, 2))
```

```python
numbers = [1, 2, 3, 4]
# map will apply a function to all the items of a list
l_product = map(lambda x: x * 2, numbers)
print(list(l_product))
```

```python
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))
```

```python
# working of reduce
# reduce(function, [a, b, c, d])
# Step 1 → function(a, b)
# Step 2 → function(result, c)
# Step 3 → function(result, d)
# Final → single value
```

```python
from functools import reduce

sum = reduce(lambda x, y: x + y, numbers)
print(sum)
```

```python
from functools import reduce

name = ["m", "u", "j", "a", "h", "i", "d"]

f_name = reduce(lambda x, y: x + y, name)
print(f_name)
```
