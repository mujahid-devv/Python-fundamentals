```python
import asyncio


async def hello():
    print("Hello")
    await asyncio.sleep(1)
    print("World")


await hello()
```

```python
# running multiple tasks
import asyncio


async def task(name):
    print(f"Start {name}")
    await asyncio.sleep(2)
    print(f"End {name}")


async def main():
    await asyncio.gather(task("A"), task("B"), task("C"))


await main()
```

```python
import asyncio
import aiohttp

urls = [
    "https://jsonplaceholder.typicode.com/posts/1",
    "https://jsonplaceholder.typicode.com/posts/2",
    "https://jsonplaceholder.typicode.com/posts/3",
]


async def fetch(session, url):
    async with session.get(url) as response:
        data = await response.json()
        print(f"{url} → {len(data)} keys")


async def main():
    async with aiohttp.ClientSession() as session:  # ClientSession — a reusable HTTP client
        tasks = [fetch(session, url) for url in urls]
        await asyncio.gather(*tasks)


await main()
```

```python
# event loop
# The event loop runs tasks one at a time, but switches between them at every await,
#  making it feel like they're running simultaneously.
```
