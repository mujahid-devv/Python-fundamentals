```python
def count(n):
    while n >=0:
        yield n
        n -=1
gen = count(10)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
gen = (x * 2 for x in range(5))

for i in gen:
    print(i)
```

```python
# List comprehension - creates a list
list_comp = [x * x for x in range(5)]
print(list_comp)

# Generator expression - creates a generator
gen_exp = (x * x for x in range(5))
# print(gen_exp)
print(list(gen_exp))
```
