```python
import json
details = {"Name":"Mujahid","Age":22,"Role":"ASE"}
s =json.dumps(details)  # python(obj) to json
d =json.loads(s)  # json -> obj

print(s)
print(d["Name"])
```

```python
import json
raw = '{"user": {"id": 42, "tags": ["admin", "mod"]}}'
data = json.loads(raw)
# accessing json 
print(data["user"]["id"])
print(data["user"]["tags"][1])
# print(data["user"]["name"]) # error
print(data.get("name","default name")) #
print(data.get("user").get("name","default name"))
```
