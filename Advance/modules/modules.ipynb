```python
if not os.path.exists("workspace"):
    os.mkdir("workspace")
else:
    print("workspace already exists")
```

```python
content = os.listdir("workspace")
print(content)
```

```python
os.makedirs("workspace/projects/code")
```

```python
# list all the files and dirs
contents = os.listdir("workspace")

files = []
dirs = []

for item in contents:
    full_path = os.path.join("workspace", item)
    if os.path.isfile(full_path):
        files.append(item)
    elif os.path.isdir(full_path):
        dirs.append(item)

print("Files:", files)
print("Directories:", dirs)
```
