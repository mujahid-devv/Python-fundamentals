```python
def dec(func):
    func()


def hello():
    print("hello world")


dec(hello)
```

```python
def my_decorator(func):
    def wrapper():
        print("before function")
        func()
        print("after function")

    return wrapper


@my_decorator
def hi():
    print("hello world")


hi()
```

```python
def my_decorator(func):
    print("decorator running")

    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)

    print("decorator returning")
    return wrapper


@my_decorator
def add(a, b):
    return a + b


add(3, 3)
```
